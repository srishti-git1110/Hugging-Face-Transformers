```python
#tokenization - AutoTokenizer class' use to load the tokenizer associated w any model
#from_pretrained method of AutoTokenizer 

from transformers import AutoTokenizer
model_ckpt = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_ckpt)
```

```python
tokenizer
```

```python
encoded_text = tokenizer('She is walking')
encoded_text
```

```python
tokens = tokenizer.convert_ids_to_tokens(encoded_text.input_ids)
tokens
```

```python
encoded_text = tokenizer('connection')
tokens = tokenizer.convert_ids_to_tokens(encoded_text.input_ids)
tokens
```

When using pretrained models, it is ***really*** important to make sure
that you use the same tokenizer that the model was trained with.
From the model’s perspective, switching the tokenizer is like shuf‐
fling the vocabulary.
